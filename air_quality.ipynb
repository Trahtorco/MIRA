```python
import numpy as np
import pandas as pd
import pytz
from pytz import timezone
import matplotlib.pyplot as plt
import matplotlib.dates as mdates                                                                                   
```

```python
data = pd.read_csv("./air_quality_data.csv", parse_dates=['created_at'], index_col="created_at", usecols=[0, 1])

fig, ax = plt.subplots()
ax.plot(data, 'r.')
ax.plot(data, 'k')
ax.xaxis.set_major_formatter(mdates.DateFormatter('%m/%d/%y %H:%M', timezone('America/Los_Angeles'))) 
plt.xticks(rotation=45)
ax.set_xlabel('Date/Time (PT)')
ax.set_ylabel(r'$>0.3\ \mu$m/dl')
plt.show()
```
